## 安装缺失依赖（与 devcontainer features 一致）

若某包未安装，`requireNamespace` 会返回 `FALSE`，下面仅对缺失包调用 `install.packages`。

```r
required_pkgs <- c(
  "languageserver",
  "quantmod",
  "xts",
  "PerformanceAnalytics",
  "dtplyr",
  "data.table"
)

missing <- required_pkgs[!sapply(required_pkgs, function(p) requireNamespace(p, quietly = TRUE))]
if (length(missing)) {
  install.packages(missing, dependencies = TRUE)
} else {
  message("All required packages are already installed.")
}

invisible(lapply(required_pkgs, function(p) {
  suppressPackageStartupMessages(library(p, character.only = TRUE))
}))
message("Loaded: ", paste(required_pkgs, collapse = ", "))
```

## （可选）注册 IRkernel，使 Jupyter 能选到 R 内核

在 **binder/rocker** 镜像中通常已注册。若在自建 R 中需：
`install.packages("IRkernel")` 后 `IRkernel::installspec()`。

```r
if (!requireNamespace("IRkernel", quietly = TRUE)) {
  install.packages("IRkernel", dependencies = TRUE)
}

# 仅在当前用户 Jupyter 未列出 R 内核时需要执行（会写用户配置）
# IRkernel::installspec(user = TRUE)

cat("IRkernel installed:", requireNamespace("IRkernel", quietly = TRUE), "\n")
```
